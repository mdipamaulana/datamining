#  JUMLAH LEMAK BERDASARKAN KATEGORI

```python
models = df.groupby('Category').count()[['Total Fat (g)']].sort_values(by='Total Fat (g)',ascending=True).reset_index()
models = models.rename(columns={'Total Fat (g)':'numberOfFat'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['Category'], y=models['numberOfFat'], color='royalblue')
plt.xticks(rotation=60)
```

# jumlah lemak

```python
engine = df.groupby('Trans Fat (g)').count()[['Sugars (g)']].sort_values(by='Sugars (g)').reset_index()
engine = engine.rename(columns={'Sugars (g)':'count'})
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=engine['Trans Fat (g)'], y=engine['count'], color='royalblue')
```

# DISTRIBUSI PROTEIN

```python
plt.figure(figsize=(15,5))
sns.distplot(df['Protein (g)'])
```

# DISTRIBUSI

```python
features = ['Protein (g)', 'Sugars (g)', 'Calories', 'Sodium (mg)', 'Serving Size (g)', 'Cholesterol (mg)', 'Carbohydrates (g)']
x = df[features]
y = df['Total Fat (g)']
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

```python
#Protein (g) = 21, Sugars (g) = 18, Calories = 580, Sodium (mg) = 1260, Serving Size (g) = 208, Cholesterol (mg) = 60, Carbohydrates (g) =54
input_data = np.array([[21,18,580,1260,208,60,54]])

prediction = lr.predict(input_data)
print('Estimasi total lemak :', prediction)
```
